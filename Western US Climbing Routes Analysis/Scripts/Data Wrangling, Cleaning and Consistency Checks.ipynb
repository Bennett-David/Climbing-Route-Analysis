##  2: Data Wrangling and Cleaning

```python
#Rename Area Latitude and Area Longitude columns- drop 'Area'
df_routes.rename(columns = {'Area Latitude' : 'Latitude', 'Area Longitude' : 'Longitude'}, inplace = True)
```

```python
# combine lat/long into new column, 'Coorinates'
df_routes['Coordinates'] = df_routes.apply(lambda row: (str(row.Latitude),str(row.Longitude)),axis=1)
print(df_routes.Coordinates[0])
```

```python
#split location column to isolate State
loc_split_1 = df_routes.Location.str.rsplit('>', 1, expand=True).rename(lambda x: f'col{x + 1}', axis=1)
```

```python
#split location column to isolate region
loc_split_2 = loc_split_1.col1.str.rsplit('>', 1, expand=True).rename(lambda x: f'col{x + 1}', axis=1)
```

```python
#split location column to isolate area
loc_split_3 = loc_split_2.col1.str.rsplit('>', 1, expand=True).rename(lambda x: f'col{x + 1}', axis=1)
```

```python
#create new column 'Area'
df_routes['Area'] = loc_split_3['col2']
```

```python
#create new column 'Region'
df_routes['Region'] = loc_split_2['col2']
```

```python
#create new column 'state'
df_routes['State'] = loc_split_1['col2']
```

```python
#New columns are all added- Coordinates, Area, Region, State
df_routes.head()
```

```python
#Drop 'your stars' column- not needed
df_routes = df_routes.drop(columns = ['Your Stars'])
```

```python
df_routes.head()
```

```python
#Rename column 'Route' to 'Route_Name'
df_routes.rename(columns = {'Route' : 'Route_Name'}, inplace = True)
```

```python
#Rename column 'Route Type' to 'Route_Type'
df_routes.rename(columns = {'Route Type' : 'Route_Type'}, inplace = True)
```

```python
df_routes.head()
```

```python
#split Route Type column to isolate subtype 
route_type_split = df_routes.Route_Type.str.rsplit(',', 1, expand=True).rename(lambda x: f'col{x + 1}', axis=1)
```

```python
route_type_split
```

```python
#create new column 'Route_subtype'
df_routes['Route_Subtype'] = route_type_split['col2']
```

```python
#Remove subtype from Route_Type column
df_routes['Route_Type'] = route_type_split['col1']
```

```python
df_routes.head()
```

```python
#re-order columns
routes_df = df_routes[['Route_Name', 'Location', 'Area', 'Region', 'State', 'Avg Stars', 'Route_Type', 'Route_Subtype', 'Rating', 'Pitches', 'Length', 'Latitude', 'Longitude', 'Coordinates', 'URL']]
```

## 3: Consistency Checks

```python
#Data stats
routes_df.describe()
```

```python
# of rows and columns
routes_df.shape
```

```python
#look for mixed type columns
for col in routes_df.columns.tolist():
  weird = (routes_df[[col]].applymap(type) != routes_df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (routes_df[weird]) > 0:
    print (col)
```

```python
#Change route_name, area, region, route_subtype, rating columns to str
routes_df = routes_df.astype({'Route_Name': 'str', 'Area': 'str', 'Region': 'str', 'Route_Subtype': 'str', 'Rating': 'str'})
```

```python
#no mixed type columns now
for col in routes_df.columns.tolist():
  weird = (routes_df[[col]].applymap(type) != routes_df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (routes_df[weird]) > 0:
    print (col)
```
